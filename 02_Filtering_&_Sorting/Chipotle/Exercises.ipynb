### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep="\t")
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
#first remove $ from item_price
prices = []
for i in chipo["item_price"]:
    prices.append(float(i[1:-1]))
chipo["clean_item_prices"] = prices

#filter for only those with quantity = 1
unique_quantities = chipo[chipo["quantity"] == 1]
unique_quantities = unique_quantities[unique_quantities["clean_item_prices"] > 10]
unique_quantities = unique_quantities.groupby(["item_name"]).min()
unique_quantities = unique_quantities["clean_item_prices"]
# print(type(unique_quantities["clean_item_prices"]))

new_df = pd.DataFrame(unique_quantities)
new_df.head()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
# divide the item_price & quntity , to get unique prices
chipo["individual_price"] = chipo["clean_item_prices"] / chipo["quantity"]
chipo.groupby(["item_name","individual_price"]).count()
```

### Step 6. Sort by the name of the item

```python
new_df.sort_values('item_name',axis = 0,inplace=True)
new_df.head()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(["clean_item_prices","quantity"],ascending = False, inplace=True)
chipo.quantity.head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo[chipo["item_name"] == "Veggie Salad Bowl"].count()
```

### Step 9. How many times did someone order more than one Canned Soda?
