### Step 9. Which was the most-ordered item? 

```python
grouped = chipo.groupby("item_name").sum()
grouped.sort_values("quantity",ascending=False,inplace=True)
grouped.head(1)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
grouped.quantity[0]
```

### Step 11. What was the most ordered item in the choice_description column?

```python
most_ordered_by_desc = chipo.groupby("choice_description").sum()
most_ordered_by_desc.sort_values("quantity", ascending = False, inplace=True)
most_ordered_by_desc.head(1)
```

### Step 12. How many items were orderd in total?

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo.info()
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo["item_price"] = chipo["item_price"].apply(lambda x: x[1::])
chipo["item_price"] = chipo["item_price"].str.strip().astype('float')

chipo.head()
```

#### Step 13.c. Check the item price type

```python
chipo.item_price.dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
# we make a revenue column for each item_price. other way, we make a function that multiplies and sums

chipo["revenue"] = chipo["item_price"] * chipo["quantity"]
chipo["revenue"].sum()
```

### Step 15. How many orders were made in the period?

```python
# len(chipo["order_id"].unique())

# unique_values = chipo.groupby(["item_name"]).nunique()
# unique_values.sort_values(["order_id"], ascending = False, inplace=True)
# unique_values.iloc[0]

organized = chipo.groupby(["item_name"]).agg(['min', 'max', 'nunique', 'count'])
organized
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1

rev_order = chipo.groupby(["order_id"]).sum()
avg_rev = rev_order["revenue"].mean()
avg_rev
```

```python
# Solution 2
rev_order = chipo.groupby(["order_id"]).agg(['sum'])
rev_order["revenue"].mean()
```
